```python
# 1. Hello World
print("1. Hello World")
print("Hello, World!")
# 2. Add Two Numbers
print("\n2. Add Two Numbers")
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Sum =", a + b)
# 3. Even or Odd
print("\n3. Even or Odd")
num = int(input("Enter a number: "))
if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")

# 4. Largest of Two Numbers
print("\n4. Largest of Two Numbers")
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
if a > b:
    print("Largest =", a)
else:
    print("Largest =", b)

# 5. Factorial
print("\n5. Factorial")
num = int(input("Enter a number: "))
fact = 1
for i in range(1, num + 1):
    fact *= i
print("Factorial =", fact)

# -----------------------------
# 6. Multiplication Table
print("\n6. Multiplication Table")
num = int(input("Enter a number: "))
for i in range(1, 11):
    print(num, "x", i, "=", num * i)

# -----------------------------
# 7. Prime Number
print("\n7. Prime Number")
num = int(input("Enter a number: "))
count = 0
for i in range(1, num + 1):
    if num % i == 0:
        count += 1
if count == 2:
    print("Prime Number")
else:
    print("Not Prime Number")
    
# 8. Reverse a String
print("\n8. Reverse a String")
text = input("Enter a string: ")
print("Reverse =", text[::-1])

# 9. Maximum in a List
print("\n9. Maximum in a List")
numbers = [10, 25, 8, 40, 15]
print("List =", numbers)
print("Largest Number =", max(numbers))

# 10. Fibonacci Series
print("\n10. Fibonacci Series")
n = int(input("Enter number of terms: "))
a = 0
b = 1
for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
print()
```
